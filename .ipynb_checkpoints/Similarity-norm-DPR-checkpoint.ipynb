```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open('DPR.output.txt')

all_events = []

for line in f:
    contents = line.split()
    all_events.append([int(contents[0])-1,int(contents[1])-1])

f.close()
```

```python
new_events = all_events[1500000:]
events = all_events[:1500000]
```

```python
len(events), len(new_events)
```

```python
users = []
machines = []
for event in events:
    users.append(event[0])
    machines.append(event[1])
```

```python
N = max(users)+1
M = max(machines)+1
access = np.array([[0.0 for j in xrange(M+1)] for i in xrange(N+1)])

print N, M
```

```python
for event in events:
    access[event[0],event[1]] += 1
```

```python
access
```

```python
for i in xrange(N+1):
    temp = np.sum(access[i,:-1])
    access[i,M] = temp

for j in xrange(M+1):
    access[N,j] = np.sum(access[:-1,j])
```

```python
access
```

## Create the profile vectors (% of accesses)

```python
profile = np.array([[0.0 for j in xrange(M+1)] for i in xrange(N+1)])
```

```python
for n in xrange(N+1):
    den = access[n,M]
    for m in xrange(M+1):
        profile[n,m] = access[n,m]/den
```

```python
profile
```

```python
profile[1,:]
```

```python
sim = np.dot(profile[1,:-1], profile[2,:-1])/(np.linalg.norm(profile[1,:-1])*np.linalg.norm(profile[2,:-1]))
sim
```

```python
sim = np.array([[0.0 for j in xrange(N)] for i in xrange(N)])
```

```python
for u in xrange(N):
    for v in xrange(N):
        sim[u,v] = np.dot(profile[u,:-1], profile[v,:-1])/(np.linalg.norm(profile[u,:-1])*np.linalg.norm(profile[v,:-1]))
```

```python
sim
```

```python
def factor(u,m):
    factor = np.dot(sim[u,:], profile[:-1,m])/np.sum(sim[u,:])
    return factor
```

```python
all_factors = []
count_new = 0
for event in new_events:
    if event[0]<N and event[1]<M:
        all_factors.append(factor(event[0],event[1]))
    else:
        all_factors.append(-1)
        count_new += 1
print 'new user/machine - '+ str(count_new)
```

```python
def low_factors(factors,cut):
    factors_low = []
    for e in factors:
        if e<=cut:
            factors_low.append(e)
    print '#entries - '+str(len(factors_low))
    return factors_low
```

```python
def high_factors(factors,cut):
    factors_high = []
    for e in factors:
        if e>cut:
            factors_high.append(e)
    print '#entries - '+str(len(factors_high))
    return factors_high
```

```python
factors = high_factors(all_factors,-1)
```

```python
def breaks(factors, n, frac):
    breaks = []
    dataList = factors[:]
    diff = []
    dataList.sort()
    for i in xrange(int(math.ceil(len(dataList)*frac)-1)):
        diff.append(dataList[i+1] - dataList[i])
    for i in xrange(n-1):
        val = max(diff)
        ind = diff.index(val)
        diff[ind] = -1
        breaks.append(dataList[ind])
    return breaks
```

```python
def getJenksBreaks( factors, numClass ): 
    dataList = factors[:]
    dataList.sort()
    mat1 = []
    for i in range(0,len(dataList)+1):
        temp = []
        for j in range(0,numClass+1):
            temp.append(0)
        mat1.append(temp)
        
    mat2 = []
    for i in range(0,len(dataList)+1):
       temp = []
       for j in range(0,numClass+1):
             temp.append(0)
       mat2.append(temp)
                
    for i in range(1,numClass+1):
        mat1[1][i] = 1
        mat2[1][i] = 0
        for j in range(2,len(dataList)+1):
            mat2[j][i] = float('inf')
            
            
    v = 0.0
    for l in range(2,len(dataList)+1):
        s1 = 0.0
        s2 = 0.0
        w = 0.0
        for m in range(1,l+1):
            i3 = l - m + 1
            
            val = float(dataList[i3-1])
            
            s2 += val * val
            s1 += val
            
            w += 1
            v = s2 - (s1 * s1) / w
            i4 = i3 - 1
            
            if i4 != 0:
               for j in range(2,numClass+1):
                    if mat2[l][j] >= (v + mat2[i4][j - 1]): 
                        mat1[l][j] = i3
                        mat2[l][j] = v + mat2[i4][j - 1]
        mat1[l][1] = 1
        mat2[l][1] = v
    
    k = len(dataList)
    kclass = []
    for i in range(0,numClass+1):
        kclass.append(0)
        
    kclass[numClass] = float(dataList[len(dataList) - 1])
    
    countNum = numClass
    while countNum >= 2:
        #print "rank = " + str(mat1[k][countNum]) 
        id = int((mat1[k][countNum]) - 2)
        #print "val = " + str(dataList[id])
        kclass[countNum - 1] = dataList[id] 
        k = int((mat1[k][countNum] - 1)) 
        countNum -= 1
    return kclass
```
