```python
users = []
machines = []
for event in events:
    users.append(event[0])
    machines.append(event[1])
```

## Build the '# accesses' matrix

```python
N = max(users)+1
M = max(machines)+1
access = np.array([[0.0 for j in xrange(M+1)] for i in xrange(N+1)])

print N, M
```

```python
for event in events:
    access[event[0],event[1]] += 1
```

```python
access
```

```python
for i in xrange(N+1):
    temp = np.sum(access[i,:-1])
    access[i,M] = temp

for j in xrange(M+1):
    access[N,j] = np.sum(access[:-1,j])
```

```python
access
```

```python
for u in xrange(N):
    if access[u,M] == 0:
        access[u,M] = -1
        
for m in xrange(M):
    if access[N,m] == 0:
        access[N,m] = -1
```

```python
profile_u = np.array([[0.0 for j in xrange(M)] for i in xrange(N)])
for u in xrange(N):
    for m in xrange(M):
        profile_u[u,m] = access[u,m]/access[u,M]
        
profile_m = np.array([[0.0 for j in xrange(M)] for i in xrange(N)])
for u in xrange(N):
    for m in xrange(M):
        profile_m[u,m] = access[u,m]/access[N,m]
```

## Compute the RP for all user pairs

```python
RP = np.matrix([[0.0 for j in xrange(N)] for i in xrange(N)])
```

```python
vals = []
for u in xrange(N):
    for v in xrange(N):
        val = np.dot(profile_u[u], profile_m[v])
        RP[u,v] = val
        vals.append(val)
```

```python
RP
```

## Compute the factors

```python
def factor(u,m):
    factor = np.dot(float(RP[u,:],profile_u[:,m]))
    return factor
```

## Threshold - Jenks natural breaks

```python
def getJenksBreaks( factors, numClass ): 
    dataList = factors[:]
    dataList.sort()
    mat1 = []
    for i in range(0,len(dataList)+1):
        temp = []
        for j in range(0,numClass+1):
            temp.append(0)
        mat1.append(temp)
        
    mat2 = []
    for i in range(0,len(dataList)+1):
       temp = []
       for j in range(0,numClass+1):
             temp.append(0)
       mat2.append(temp)
                
    for i in range(1,numClass+1):
        mat1[1][i] = 1
        mat2[1][i] = 0
        for j in range(2,len(dataList)+1):
            mat2[j][i] = float('inf')
            
            
    v = 0.0
    for l in range(2,len(dataList)+1):
        s1 = 0.0
        s2 = 0.0
        w = 0.0
        for m in range(1,l+1):
            i3 = l - m + 1
            
            val = float(dataList[i3-1])
            
            s2 += val * val
            s1 += val
            
            w += 1
            v = s2 - (s1 * s1) / w
            i4 = i3 - 1
            
            if i4 != 0:
               for j in range(2,numClass+1):
                    if mat2[l][j] >= (v + mat2[i4][j - 1]): 
                        mat1[l][j] = i3
                        mat2[l][j] = v + mat2[i4][j - 1]
        mat1[l][1] = 1
        mat2[l][1] = v
    
    k = len(dataList)
    kclass = []
    for i in range(0,numClass+1):
        kclass.append(0)
        
    kclass[numClass] = float(dataList[len(dataList) - 1])
    
    countNum = numClass
    while countNum >= 2:
        #print "rank = " + str(mat1[k][countNum]) 
        id = int((mat1[k][countNum]) - 2)
        #print "val = " + str(dataList[id])
        kclass[countNum - 1] = dataList[id] 
        k = int((mat1[k][countNum] - 1)) 
        countNum -= 1
    return kclass
```

```python
def breaks(factors, n):
    breaks = []
    dataList = factors[:]
    diff = []
    #dataList.sort()
    for i in xrange(len(dataList)-1):
        diff.append(dataList[i+1] - dataList[i])
    for i in xrange(n-1):
        val = max(diff)
        ind = diff.index(val)
        diff.remove(val)
        breaks.append(dataList[ind])
    return breaks
```
