```python
f = open('cbs.output.txt')

all_events = []

for line in f:
    #print line
    contents = line.split()
    #print contents
    all_events.append([int(contents[0])-1,int(contents[1])-1])

f.close()
```

```python
new_events = all_events[10000000:]
events = all_events[:10000000]
```

```python
len(events)
```

```python
len(new_events)
```

```python
users = []
machines = []
for event in events:
    users.append(event[0])
    machines.append(event[1])
```

## Build the '# accesses' matrix

```python
N = max(users)+1
M = max(machines)+1
access = np.matrix([[0.0 for j in xrange(M+1)] for i in xrange(N+1)])

print N, M
```

```python
for event in events:
    access[event[0],event[1]] += 1
```

```python
access
```

```python
for i in xrange(N+1):
    temp = np.sum(access[i,:-1])
    access[i,M] = temp

for j in xrange(M+1):
    access[N,j] = np.sum(access[:-1,j])
```

```python
access
```

```python
for u in xrange(N):
    if access[u,M] == 0:
        access[u,M] = -1
        
for m in xrange(M):
    if access[N,m] == 0:
        access[N,m] = -1
```

## Compute the RP for all user pairs

```python
RP = np.matrix([[0.0 for j in xrange(N)] for i in xrange(N)])
```

```python
for u in xrange(N):
    tot = access[u,M]
    for v in xrange(N):
        for m in xrange(M):
            RP[u,v] += (access[u,m])*(access[v,m]/access[N,m])
        RP[u,v] = RP[u,v]/tot
```

```python
RP
```

## Compute the factors

```python
def factor(u,m):
    factor = 0
    for v in xrange(N):
        factor += RP[u,v]*access[v,m]/access[v,M]
    return factor
```
