```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open('synthetic.output.txt')

all_events = []

for line in f:
    contents = line.split()
    all_events.append([int(contents[0])-1,int(contents[1])-1])

f.close()
```

```python
new_events = all_events[100000:]
events = all_events[:100000]
```

```python
len(events), len(new_events)
```

```python
users = []
machines = []
for event in events:
    users.append(event[0])
    machines.append(event[1])
```

```python
global N, M
N = max(users)+1
M = max(machines)+1
profile = np.array([[0.0 for j in xrange(M+1)] for i in xrange(N+1)])

print N, M
```

```python
for event in events:
    profile[event[0],event[1]] += 1
```

```python
profile
```

```python
for i in xrange(N+1):
    temp = np.sum(profile[i,:-1])
    profile[i,M] = temp

for j in xrange(M+1):
    profile[N,j] = np.sum(profile[:-1,j])
```

```python
profile
```

```python
global profile
```

```python
for n in xrange(N):
    den = profile[n,M]
    for m in xrange(M):
        profile[n,m] = profile[n,m]/den
```

```python
profile
```

```python
global sim
sim = np.array([[0.0 for j in xrange(N)] for i in xrange(N)])
```

```python
for u in xrange(N):
    for v in xrange(u+1):
        val = np.dot(profile[u,:-1], profile[v,:-1])/(np.linalg.norm(profile[u,:-1])*np.linalg.norm(profile[v,:-1]))
        sim[u,v] = val
        sim[v,u] = val
```

```python
sim
```

```python
def update(event):
    u = event[0]
    m = event[1]
    T = profile[u,M]+1
    f = (T-1)/T
    for i in xrange(M):
        profile[u,i] = profile[u,i]*f
    profile[u,m] += 1/T
    profile[u,M] = T
    
    for v in xrange(N):
        val = np.dot(profile[u,:-1], profile[v,:-1])/(np.linalg.norm(profile[u,:-1])*np.linalg.norm(profile[v,:-1]))
        sim[u,v] = val
        sim[v,u] = val
```

```python
good_access = []
bad_access = []

def pos_feed(good_access, event):
    good_access.append(event)
    
def neg_feed(bad_access, event):
    bad_access.append(event)
    u = event[0]
    m = event[1]
    bad = profile[u,m]*profile[u,M]
    T = profile[u,M] - bad
    f = (T+bad)/T
    for i in xrange(M):
        profile[u,i] = profile[u,i]*f
    profile[u,m] = 0
    profile[u,M] = T
    
    for v in xrange(N):
        val = np.dot(profile[u,:-1], profile[v,:-1])/(np.linalg.norm(profile[u,:-1])*np.linalg.norm(profile[v,:-1]))
        sim[u,v] = val
        sim[v,u] = val
```

```python
def score(event, good_access, bad_access):
    if event[0]>=N or event[1]>=M:
        return 'new'
    if event in good_access:
        return 'good'
    if event in bad_access:
        return 'bad'
    else:
        u = event[0]
        m = event[1]
        score = np.dot(sim[u,:], profile[:-1,m])
        return score
```

```python
factors = []
i = 0
for event in new_events:
    i += 1
    if i%1000 == 1:
        print i
    #update(event)
    factors.append(score(event, good_access, bad_access))
```

```python
scores = []
i = 0
for event in new_events:
    i += 1
    if i%1000 == 1:
        print i
    update(event)
    scores.append(score(event, good_access, bad_access))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Calculate the point density
xy = np.vstack([[i for i in xrange(len(scores))], factors])
z = gaussian_kde(xy)(xy)

fig, ax = plt.subplots()
ax.scatter([i for i in xrange(len(scores))], factors, c=z, s=20, edgecolor='')
plt.xlabel('factor')
plt.ylabel('event')
```
